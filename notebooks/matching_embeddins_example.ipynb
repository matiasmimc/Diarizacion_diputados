## Cargar archivos
De los embeddings ya etiquetados y el resultado de la predicción de turnos (obtenida con Diarizador)

```python
import json
embs_file = "/content/drive/My Drive/Diarizacion/embeddings_por_mpuid.json"
with open(embs_file, "r", encoding="utf-8") as f:
    embs_dict = json.load(f)

from diarizacion_pipeline.utils import load_embeddings

embs_dict = load_embeddings(embs_dict)

import pickle
segments_file = "/content/drive/MyDrive/Diarizacion/Segmentos_whisperx/segments_raw70.pkl"
with open(segments_file, "rb") as f:
    all_segments = pickle.load(f)
```

## Inicializar clase de emparejamiento de speakers.

```python
from diarizacion_pipeline.Matcher_speakers import Matcher_speakers

matcher = Matcher_speakers(all_segments, embs_dict)
```

## Matchear
Se puede especificar un threshold de similaridad coseno entre embeddings. Es $0.75$ por defecto.

```python
transcripcion = matcher.match(sim_threshold=0.75)
transcripcion
```
